```python
import os

def print_tree(path, ignore_ext='.png', indent=''):
    for item in sorted(os.listdir(path)):
        item_path = os.path.join(path, item)
        if os.path.isdir(item_path):
            if item == '.git':
                continue
            print(f"{indent}📁 {item}")
            print_tree(item_path, ignore_ext, indent + '    ')
        elif not item.endswith((ignore_ext, '.jpg')):
            print(f"{indent}📄 {item}")

# Replace '/path/to/directory' with the directory you want to print
print_tree('.')
```
